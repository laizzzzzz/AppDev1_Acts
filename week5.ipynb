```python
import torch
from transformers import pipeline
import os
```

```python
# load model
dolly_pipeline = pipeline(model="databricks/dolly-v2-3b", torch_dtype=torch.bfloat16, trust_remote_code=True, device_map="auto")
```

```python
#helper function
def get_completion_dolly(Input):
  system = F"""
  You are the best among the Rest!
  You are good at explaining everything about the world.
  Help the needy as much as you can.
  """

  prompt = f"System: {system}\nUser: {input}\nResponse from Dolly:"
  print(prompt)

  dolly_response = dolly_pipeline(prompt, max_new_tokens=500)

  return dolly_response[0]["generated_text"]
```

```python
prompt = "What is the meaning of Life?"
print(get_completion_dolly(prompt))
```

```python
import gradio as gr

interface = gr.Interface(fn=get_completion_dolly, inputs=[gr.Textbox(label="Type your prompt here", lines=6)],
                         outputs=[gr.Textbox(label="Type your answer here", lines=3)],
                         title="D'BestBot",
                         examples=["What is the meaning of the word Life?" , "What is the study of about life on earth?"])

interface.launch(share=True, debug=True)
```
